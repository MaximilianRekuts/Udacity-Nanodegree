```python
#load required libraries 
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

```python
#load csv file
df1 = pd.read_csv('/Users/maximilianrekuts/Desktop/results.csv', encoding = 'utf-8') 
```

```python
#look at column
df1.columns
```

```python
#check for missing values
print(df1.isnull().sum())
```

```python
#split dataframe into two of which one hold local weather and the other global weather
lt = df1.iloc[:,[0,3]]
gt = df1.iloc[:,[0,1]]
```

```python
# set year column as index for local and global temperature 
lt.set_index('year', inplace=True)
gt.set_index('year', inplace =True)
```

```python
# line plot - the yearly average weather temperature in Berlin and World
fig= plt.figure(figsize=(25,15))
plt.plot(lt,'red')
plt.plot(gt,'blue')
plt.legend(['Berlin','Global'], fontsize=16)
plt.xlabel('year', fontsize=20)
plt.ylabel('temperature', fontsize=20)
plt.title('The average temperature in Berlin and World', fontsize=25)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
```

```python
# Berlin weather temperature 
# the simple moving average over a period of 10, 20, 30 years
lt_sma1 = lt.rolling(10, min_periods=10).mean()
lt_sma2 = lt.rolling(20, min_periods=20).mean()
lt_sma3 = lt.rolling(30, min_periods=30).mean()
```

```python
# line plot - the yearly average air temperature in Berlin
fig= plt.figure(figsize=(25,15))
plt.plot(lt_sma1, 'sandybrown')
plt.plot(lt_sma2, 'grey')
plt.plot(lt_sma3, 'green')
plt.legend(['SMA1','SMA2','SMA3'], fontsize=16)
plt.xlabel('year', fontsize=20)
plt.ylabel('temperature in C', fontsize=20)
plt.title('The 10, 20, 30 years moving average (SMA) weather temperature in Berlin', fontsize=25)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)

```

```python
# World weather temperature 
# the simple moving average over a period of 10, 20, 30 years
gt_sma1 = gt.rolling(10, min_periods=10).mean()
gt_sma2 = gt.rolling(20, min_periods=20).mean()
gt_sma3 = gt.rolling(30, min_periods=30).mean()
```

```python
#plot graph for simple moving average over a period of 10, 20, 30 years for world weather temperature 
fig= plt.figure(figsize=(25,15))
plt.plot(gt_sma1, 'sandybrown')
plt.plot(gt_sma2, 'grey')
plt.plot(gt_sma3, 'green')
plt.legend(['SMA1','SMA2','SMA3'], fontsize=16)
plt.xlabel('year', fontsize=20)
plt.ylabel('temperature in C', fontsize=20)
plt.title('The 10, 20, 30 years simple moving average (SMA) weather temperature in the world', fontsize=25)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
```

```python
fig= plt.figure(figsize=(25,15))
plt.plot(lt_sma1, 'red')
plt.plot(gt_sma1, 'blue')
plt.legend(['SMA1-Berlin','SMA1-World'], fontsize=16)
plt.xlabel('year', fontsize=20)
plt.ylabel('temperature in C', fontsize=20)
plt.title('The 10 years simple moving average (SMA) weather temperature in Berlin and World', fontsize=25)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
```
